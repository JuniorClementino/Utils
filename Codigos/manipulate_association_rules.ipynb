```python
#Reading file
df = pd.read_csv("sem_ponderar.csv") 
```

```python
#Removing column Target
df =df.drop(columns=['Unnamed: 0']) 
df.head()        
```

```python
#Transforming int to string
for column in df:
    df[column] = df[column].astype(str)
    #print(df[column])
    
df.dtypes

   
    
```

```python
list_target = df['target'].tolist()
```

```python

df =df.drop(columns=['target']) 
```

```python
#Change value "1" in table to column names
names = df.columns.values

_i = 0
for col in df:
    for i, row_value in df[col].iteritems():        
        if row_value == '1' or row_value == '2':
            df.iloc[i,_i] = names[_i]
            
    _i += 1 
```
